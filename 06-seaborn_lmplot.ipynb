## สร้างกราฟด้วย seaborn Linear Regression
## seaborn (lmplot => regplot + Facegrid)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
url='https://raw.githubusercontent.com/PangPriya13/tutorial/main/mpg.csv'
df=pd.read_csv(url)
df.head()
```

```python
df['gear']=df.trans.str[0].map({'a':'auto','m':'manual'}).astype("category")
df['drv']=df.drv.map({'f':'front','r':'rear','4':'4-wheel'}).astype("category")
df.head()
```

```python
sns.lmplot(data=df, x='displ', y='cty');
```

```python
sns.lmplot(data=df, x='displ', y='cty',
          fit_reg=False);
```

```python
sns.lmplot(data=df, x='displ', y='cty', hue='class');
```

```python
sns.lmplot(data=df, x='displ', y='cty', hue='gear');
```

```python
sns.lmplot(data=df, x='displ', y='cty',
           hue='gear', ci=False);
```

```python
sns.lmplot(data=df, x='displ', y='cty',
          hue='drv');
```

```python
sns.lmplot(data=df, x='displ', y='cty',
          hue='drv',
          col='gear', height=5,
          scatter_kws={'alpha':0.5});
```

```python
sns.lmplot(data=df, x='displ', y='cty',
          hue='drv',
          col='gear',
           height=4,
          scatter=False,
          ci=False);
```
